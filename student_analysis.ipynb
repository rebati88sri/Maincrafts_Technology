```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading Data 


```python
df = pd.read_csv("student-mat.csv", sep=";")
print(df.head(5))
```

loading the csv data which is ; separated

### Check Missing Values

```python
print(df.isnull().sum())
```

### Remove Duplicate Rows

Removing unnecessary Duplicate Rows from data set to improve prediction 

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

### Average Final Grade (G3)

finding Average of grade by mean() function

```python
average_grade = df["G3"].mean()

print("Average Final Grade:", round(average_grade,2))
```

### Number of Students Scoring Above 15

finding number of student score avobe 15 mark

```python
above_15 = df[df["G3"] > 15]

print("Students scoring above 15:", len(above_15))
```

### Correlation Between Study Time and Final Grade

finding relation bw study time and final grade in percentage

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation:", correlation)
```

### Which Gender Performs Better?

```python
gender_avg = df.groupby("sex")["G3"].mean()

print(gender_avg)
```

### Histogram of Final Grades

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades")

plt.xlabel("Final Grade (G3)")

plt.ylabel("Number of Students")

plt.show()
```

### Scatter Plot (Study Time vs Final Grade)

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")

plt.xlabel("Study Time")

plt.ylabel("Final Grade (G3)")

plt.show()
```

### Bar Chart (Average Score by Gender)

```python
plt.figure(figsize=(6,5))

gender_avg.plot(kind="bar")

plt.title("Average Final Grade by Gender")

plt.xlabel("Gender")

plt.ylabel("Average G3")

plt.show()
```
